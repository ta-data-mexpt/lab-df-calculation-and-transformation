## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
amount_sum = orders.groupby('CustomerID', as_index=False).agg({'amount_spent':'sum'})
```

```python
amount_sum
```

```python
vip= amount_sum[amount_sum['amount_spent'] > amount_sum['amount_spent'].quantile(0.95)]
vip['type'] = 'vip'

vip = vip[['CustomerID', 'type']]
```

```python
preferred=amount_sum[(amount_sum['amount_spent'] <= amount_sum['amount_spent'].quantile(0.95)) & 
           (amount_sum['amount_spent'] >= amount_sum['amount_spent'].quantile(0.75))]
preferred['type'] = 'preferred'
```

```python
preferred = preferred[['CustomerID', 'type']]
```

```python
conditions = [
    (amount_sum['amount_spent'] <= amount_sum['amount_spent'].quantile(0.95)) & 
    (amount_sum['amount_spent'] >= amount_sum['amount_spent'].quantile(0.75)),
    (amount_sum['amount_spent'] > amount_sum['amount_spent'].quantile(0.95)),
    (amount_sum['amount_spent'] < amount_sum['amount_spent'].quantile(0.75))
    ]

values = ['preferred', 'vip','regular']
```

```python
amount_sum['type'] = np.select(conditions, values)
```

```python
amount_sum= amount_sum[['CustomerID', 'type']]
```

```python
class_orders= pd.merge(orders,amount_sum, on= ['CustomerID'])
```

```python
class_orders.head(10)
```
