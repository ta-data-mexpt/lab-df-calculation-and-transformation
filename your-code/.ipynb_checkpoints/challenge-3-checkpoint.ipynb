```python
filtered=orders.groupby("CustomerID",as_index=False)["amount_spent"].sum()
filtered.head()
```

```python
labels=["P20","P40","P60","P80","P99","VIP"]
filtered["Category"]=pd.qcut(filtered["amount_spent"],q=[0,.2,.4,.6,.8,.99,1],labels=labels)
```

```python
filtered.head()
```

# Q2: How to identify which country has the most VIP Customers?

```python
filtered2=filtered.merge(orders[["CustomerID","Country"]],how="left",left_on="CustomerID",right_on="CustomerID")
filtered2
```

```python
filtered3=filtered2[filtered2["Category"]=="VIP"].groupby(["Country"],as_index=False)["Category"].count()
```
