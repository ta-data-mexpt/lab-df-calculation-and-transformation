```python
# your code here
col_names=['CustomerID','Total Amount']
TotalAmount = orders.groupby(['CustomerID'])['amount_spent'].sum().sort_values(ascending = False)
TotalAmount
```

```python
TotalAmount.quantile([0.25, 0.50, 0.75])
```

```python
TotalAmount['TypeClients'] = np.where(TotalAmount['amount_spent'] >= 1661.195,'VIP',
                                      np.where(TotalAmount['amount_spent'] >= 674.450, 'Preferred', 'Regular'))
TotalAmount.head(10)
```
